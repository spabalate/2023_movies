## Top 20 Movies in 2023 vs Top 20 movies of all time

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

this_year = pd.read_csv('/Users/sam/Desktop/MY_UC_BERK/PROJECTS/PROJECT_3/top_movies_2023.csv')
all_time = pd.read_csv('/Users/sam/Desktop/MY_UC_BERK/PROJECTS/PROJECT_3/goat_movies.csv')

this_year_top20 = this_year.head(20)
all_time_top20 = all_time.head(20)

# Set a custom figure size (width, height) in inches
plt.figure(figsize=(10, 10))

bar_width = 0.35  # Adjust the width of the bars

# Create double bar plot with Seaborn
sns.barplot(x=this_year_top20['Rank'], y=this_year_top20['Gross Earnings'], color='blue', label='Top 2023 Movies', alpha=0.7)
sns.barplot(x=all_time_top20['Rank'], y=all_time_top20['Lifetime Gross'], color='orange', label='Top Movies of All Time', alpha=0.7)

# Format the y-axis ticks as dollar amounts
formatter = FuncFormatter(lambda x, _: f'${int(x):,}')
plt.gca().yaxis.set_major_formatter(formatter)

# Set the background color
palette = sns.color_palette('ch:rot=-.25,hue=1,light=.75', n_colors=8)
background_color = palette[0]
opacity = 0.25
plt.gca().set_facecolor((*background_color, opacity))
plt.gca().patch.set_alpha(opacity)

# Customize the plot
plt.title('Movie Comparison')
plt.xlabel('Ranking')
plt.ylabel('Gross')

# Display movie titles as a legend
plt.legend()

plt.tight_layout()
# Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

# Load your data
this_year = pd.read_csv('/Users/sam/Desktop/MY_UC_BERK/PROJECTS/PROJECT_3/top_movies_2023.csv')
all_time = pd.read_csv('/Users/sam/Desktop/MY_UC_BERK/PROJECTS/PROJECT_3/goat_movies.csv')

# Set a custom figure size (width, height) in inches
plt.figure(figsize=(10, 6))

# Define the 'rocket_r' color palette
palette = sns.color_palette('rocket_r')

# Plot the KDE plot for the distribution of Gross for Top 2023 Movies
sns.kdeplot(data=this_year, x='Gross Earnings', fill=True, label='Top 2023 Movies', color=palette[0])

# Plot the KDE plot for the distribution of Lifetime Gross for Top Movies of All Time
sns.kdeplot(data=all_time, x='Lifetime Gross', fill=True, label='Top Movies of All Time', color=palette[3])

# Format the x-axis ticks as dollar amounts
formatter = FuncFormatter(lambda x, _: f'${int(x):,}')
plt.gca().xaxis.set_major_formatter(formatter)

# Customize the plot
plt.title('Movie Comparison')
plt.xlabel('Gross')
plt.ylabel('Ranking')

# Set the background color
palette = sns.color_palette('ch:rot=-.25,hue=1,light=.75', n_colors=8)
background_color = palette[0]
opacity = 0.25
plt.gca().set_facecolor((*background_color, opacity))
plt.gca().patch.set_alpha(opacity)

# Display movie titles as a legend
plt.legend()

plt.tight_layout()
# Show the plot
plt.show()
```

```python
# Sort DataFrame by 'Release Date'
this_year = this_year.sort_values(by='Release Date')

# Create a scatter plot
sns.scatterplot(x='Number of Theaters', y='Gross Earnings', hue='Release Date', style='Release Date', data=this_year)

# Format the x-axis ticks as dollar amounts
formatter = FuncFormatter(lambda y, _: f'${int(y):,}')
plt.gca().yaxis.set_major_formatter(formatter)

# Add labels and title
plt.xlabel('Number of Theaters')
plt.ylabel('Gross Earnings')
plt.title('Scatter Plot of Number of Theaters vs. Gross Earnings (2023)')

# Set the background color
palette = sns.color_palette('ch:rot=-.25,hue=1,light=.75', n_colors=8)
background_color = palette[0]
opacity = 0.25
plt.gca().set_facecolor((*background_color, opacity))
plt.gca().patch.set_alpha(opacity)

legend = plt.legend(title='Release Date', bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0., ncol=4)

# Show the plot
plt.show()
```
